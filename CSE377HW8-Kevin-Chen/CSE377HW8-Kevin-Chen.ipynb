```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.transform import rotate
```

### Part 01 - Using the forward projection to generate the CT signals (sinogram).

#### For example, the following is an 800 x 800 image and its sinogram 800 x 180 for 180 viewing angles (i.e., the x-ray source rotates around the head every 1 degree clockwise from 0 degree to 180 degree).

```python
# Load Image

img = cv2.imread("Thanos-1.jpg", cv2.IMREAD_GRAYSCALE)
size = img.shape[0] # Square Image
plt.imshow(img, cmap='gray')
print(img.shape)
```

```python
# Convert Image to Sinogram
angles = np.arange(0,180) # you can change the step
sinogram_img = np.zeros((size, len(angles)), dtype=np.float32) # every column is the detector signal from a view angle

for i in range(180):
    # rotate the image by counterclockwise so the x-ray is clockwise relevantly
    # projection is basically summing over columns
    sinogram_img[:,i] = rotate(img, angles[i]).sum(axis=0)

plt.imshow(sinogram_img, cmap='gray')
plt.title('sinogram')
plt.show()
```

### Part 02 - Using the backward projection to reconstruct the CT image from the sinogram.

### The reconstructed image should be blurred and have low-contrast.

```python
# Creating a function for backward projection because we need it to use this multiple times
# Rather than repeat code, I would just create the function once.

def back_project(arr):
    # The sinogram is shaped as (image size, 180 degrees) so split these values
    img_size, num_angles = arr.shape
    # Creating an empty numpy 2d array with the dimensions of the original image, which is
    # a square in this case (800,800)
    reconstructed_img = np.zeros((img_size, img_size), dtype=np.float32)
    # Need to find the derivative theta, which would simply be 1.0, since the angles are from
    # 0 to 179, inclusive, which are incremented by 1 degrees each time. i.e., 0, 1, 2, 3, ...
    dTheta = 180.0 / num_angles
    # for-loop to iterate through each angle from the range of 0 to 179.
    for i in range(num_angles):
        # creating a temporary array to store these projected values for the backward projected
        # image. Rotation to temp_arr simulates the projection of the image at that angle.
        # Adding temp arr to reconstructed_img would add up the many different angles.
        temp_arr = np.tile(arr[:,i], (img_size, 1))
        temp_arr = rotate(temp_arr, -dTheta*i)
        reconstructed_img += temp_arr
    # This step is to make sure the values of the reconstructed 2d array is between the values of 0 and 255, inclusive.
    reconstructed_img = (reconstructed_img - reconstructed_img.min()) * (255 - 0) / (reconstructed_img.max() - reconstructed_img.min()) + 0
    reconstructed_img = np.clip(reconstructed_img, 0, 256 -1). astype(int)
    return reconstructed_img
```

```python
# Calling that defined backward projection function to get the reconstructed image
# of the original sinogram from the input image.

reconstructed_bp = back_project(sinogram_img)

# Plotting the reconstructed image using the backward projection.

plt.imshow(reconstructed_bp, cmap='gray')
plt.title('Reconstruction by Back Projection')
plt.tick_params(left=False,right=False,labelleft=False,labelbottom=False,bottom=False)
plt.show()
```

### Part 03 - Histogram Equalization.

### Applying the histogram equalization to the reconstruction from part 02 will result in a better visualization.

```python
# Creating a custom histogram of the pixel intensities of the reconstructed image from
# part 02. The logic follows similarly to what was presented in assignment 04 in class.

# defining a histogram-equalizing function for this

def hist_equalize(arr):
    # creating an empty array that has length 256 for the pixel intensities [0,255].
    hist = np.zeros(256)
    # the for-loop iterates through each pixel intensity from 0 to 255, and adds up the count 
    # of intensity equaling the value present in the reconstructed image.
    for intensity in range(256):
        hist[intensity] = np.sum(arr == intensity)
    # normalizing
    hist /= (size*size)
    # finding the cumulative summation of the normalized intensity histogram
    norm_cdf = np.cumsum(hist)
    # applying the intensity transformation function: s_k = T(r_k) = (L-1) * cdf(r_k), where
    # L = 256.
    hist_img = 255 * norm_cdf[arr]
    hist_img = (np.rint(hist_img)).astype(int)
    return hist_img

# Displaying the histogram-equalized reconstructed image, which looks sharper than before.

hist_eq_img = hist_equalize(reconstructed_bp)
plt.title('Reconstruction by Back Projection after Histogram Equalization')
plt.imshow(hist_eq_img, cmap='gray')
plt.tick_params(left=False,right=False,labelleft=False,labelbottom=False,bottom=False)
plt.show()
```

### Part 04 - Frequency Filtering.

### Applying ramp filtering (a high-pass filtering) onto each column of the sinogram in the frequency domain to attenuate the low-frequency components. Leaves us with a filtered sinogram.

```python
# creating the ramp filter which is a piece-wise function, where (y-axis) 1->0 from 
# (x-axis) 0->400, and (y-axis) 0->1 from (x-axis) 400->800.

ramp = np.concatenate((np.linspace(1, 0, num=400), np.linspace(0, 1, num=400)))
plt.plot(ramp)
plt.title('Ramp Filter')
plt.show()
```

```python
# creating a function for ramp filtering.
def ramp_filter(arr):
    
    # creating an empty array to store the values of the ramp filtered sinogram image.

    filtered_arr = np.zeros_like(arr)

    # applying the ramp filter to each column of the sinogram

    for i in range(arr.shape[1]):
        
        # using fft.fftshift to get into the frequency domain of the sinogram image
        
        column = arr[:, i]
        fft = np.fft.fftshift(np.fft.fft(column))
        filtered_fft = ramp * fft
        filtered_column = np.fft.ifft(np.fft.ifftshift(filtered_fft)).real
        filtered_arr[:, i] = filtered_column
    
    return filtered_arr
```

```python
# plotting the sinogram after ramp filtering

filtered_sinogram = ramp_filter(sinogram_img)
plt.imshow(filtered_sinogram, cmap='gray')
plt.title('sinogram after ramp filtering')
plt.show()
```

### Part 05 - Filtered Back Projection.

### Performing the back projection on the filtered sinogram will result in a sharper reconstructed CT image.

```python
# calling the defined function for backward projection again, but this time on the 
# ramp filtered sinogram from the previous part, part 04.

reconstructed_fbp = back_project(filtered_sinogram)

# plotting the reconstructed image using backward projection on the ramp filtered sinogram.

plt.imshow(reconstructed_fbp, cmap='gray')
plt.title('Reconstruction by Filtered Back Projection')
plt.tick_params(left=False,right=False,labelleft=False,labelbottom=False,bottom=False)
plt.show()
```

### Part 06 

### Imagine some of the x-ray detectors are out of order, so some rows of the original sinogram are black (i.e., with zero values). Performing the filtered back projection on the damaged sinogram, what will you find?

```python
# creating a function that uses the code from previous parts of this homework
# to easily perform filtered back projection.

def filtered_back_proj(arr):
    
    # applying the ramp filter onto the input array, which would be the sinogram
    
    filtered_sinogram = ramp_filter(arr)
    reconstructed_sinogram = back_project(filtered_sinogram)
    
    return reconstructed_sinogram
```

```python
# black_rows is randomly choosing 45 rows out of 800 to turn black or have a value of 0.

black_rows = np.random.choice(np.arange(size), 45, replace=False)
black_rows_sinogram = sinogram_img.copy()
black_rows_sinogram[black_rows, :] = 0.0

# plotting this sinogram that has 45 black rows, as we can see this works.

plt.imshow(black_rows_sinogram, cmap='gray')
plt.title('sinogram')
plt.show()
```

```python
# using forward backward projection to reconstruct the sinogram with 45 black rows

part6_fbp = filtered_back_proj(black_rows_sinogram)
plt.imshow(part6_fbp, cmap='gray')
plt.title('Reconstruction by Filtered Back Projection - Missing Rows')
plt.tick_params(left=False,right=False,labelleft=False,labelbottom=False,bottom=False)
plt.show()

# As we can see, performing the filtered back projection on the damaged sinogram appears
# to not remotely look similar to what we had as a result from part 05. We notice how the
# image appears to be very gray and have a spiral pattern that is goes around the center of
# the reconstructed image. This is the result from performing filtered back projection on 
# a damaged sinogram, which has x-ray detectors out of order and in turn has missing rows 
# in the sinogram. The original image isn't remotely visible in this output image.
```

### Part 07

### Imagine some view angles of the x-ray sources are out of order, so some columns of the original sinogram are black (i.e., with zero values). Performing the filtered back projection on the damaged sinogram, what will you find?

```python
# black_cols is randomly choosing 45 columns out of the 180 to turn black,
# or have a value of 0. It is 180 because the columns of the sinogram correlate to the 
# angles of the detectors from 0 to 179, or 1 to 180 degrees.

black_cols = np.random.choice(np.arange(len(angles)), 45, replace=False)
black_cols_sinogram = sinogram_img.copy()
black_cols_sinogram[:, black_cols] = 0.0

# Plotting the sinogram with 45 black columns for viewing.

plt.imshow(black_cols_sinogram, cmap='gray')
plt.title('sinogram')
plt.show()
```

```python
# using forward backward projection to reconstruct the sinogram with 45 black columns.

part7_fbp = filtered_back_proj(black_cols_sinogram)

plt.imshow(part7_fbp, cmap='gray')
plt.title('Reconstruction by Filtered Back Projection - Missing Columns')
plt.tick_params(left=False,right=False,labelleft=False,labelbottom=False,bottom=False)
plt.show()


# After performing the filtered back projection on the damaged sinogram, which has 45 columns
# of the original sinogram become black, we can notice that the reconstructed image is very
# visible, which closely resembles the filtered back projection image from part 05. While
# the x-ray detectors being out of order causes the filtered back projection to be unrecognizable
# to the original image, the viewing angles of the x-rays out of order doesn't affect the output
# image by much. There seems to be a bit more lines being distributed in the image, but the
# overall image of Thanos appears to be crisp and sharp. 
```
